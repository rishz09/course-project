```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_aug = pd.read_feather('augmented_MNIST.feather').iloc[:, 1:].to_numpy()
```

```python
plt.imshow(x_aug[1006].reshape(28, 28), cmap='gray')
plt.axis('off')
plt.show()
```

```python
fig, axs = plt.subplots(2, 5, figsize=(12, 6))
axs = axs.ravel()

for i in range(10):
    if i < 10:
        axs[i].imshow(x_aug[1006 + 60000 * i].reshape(28, 28), cmap='gray')
        axs[i].axis('off')

plt.tight_layout()
plt.savefig('Images/image_augmentation_combinations.png')
plt.show()
```
